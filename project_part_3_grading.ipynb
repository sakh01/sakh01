#Introduction:

The goal of the paper is to design a quadrupedal robotic with high speed and flexible trunk. They designed two models; Mass-Mass-Spring (MMS) model and Spring Loaded Inverted Pendulum (SLIP) model. Result shows that MMS model quadrupedal robotic was able to make robotic run faster as it generated a minimum bending moment at the end of robotic trunk which result in storing and releasing elastic energy while running and having a flexible trunk

Q2: What are the observations? How many, and how were they collected?

The data collected by developing the MMS (Mass-Mass-Spring) model for robotics and measure
their performance in flight and stance phase, then they compared their data with the SLIP
(Spring Loaded Inverted Pendulum) model. To do that they developed a simulation setup to
measure the stored energy in trunk of robot and how mass of the leg (m) that is located in knee
joint (mass leg is only performed in the MMS model) will make a difference in the robotic speed
between SLIP and MMS models. This was done in both types of gallops (rotatory and transverse
gallop) of cheetah and horses respectively. The dataset for the MMS flight phase has 51
observations at different timepoints, each row is considered an observation. Observation is
developing quadruped robots with high speed and flexible trunk.


Q3: What are the features? How many, and what do they tell us about the observations.


Main features are the minimum bending moment and maximum bending moment in both types
of gallops (rotatory and transverse) in both models SLIP and MMS. At each time point we
measure various characteristics that related to robotic movement such as (stride length,
hopping height, normal effort, tangential stress, minimum and maximum bending moments,
bending difference, Hip mass in MMS model (M), Knee mass in MMS model, rotary gallop,
transversal gallop, stance and flight phase, MMS and SLIP models). The main features are
important because they tell us about the spin bending and whether that will make a difference
in the robotic speed. The minimum bending moment was crucial because it gives us more
information about the observation result in the MMS model in generating an accurate bending
of the trunk. In total there were 15 features.

Q4: What other data or knowledge are used?

Other data or knowledge are more related to physics like equations used in the MMS model to
represent the stance phase in horizontal and vertical axes. Where the ẍm and ẍM are horizontal
acceleration and the ÿm and ÿM are the vertical acceleration of the knee mass (m), and hip mass
(M) respectively for the stance phase. The same symbols (ẍm and ẍM and ÿm and ÿM) but with
different equations were used in the flight phase.
